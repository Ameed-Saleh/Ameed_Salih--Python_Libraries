```python
# Q6
# Allocate an uninitialized block of memory for a (4, 3) array — the fastest possible allocation. You will fill every cell before reading it.
# Goal: reserve uninitialized memory
# shape: (4, 3)  |  dtype: float64
# ⚡ Fastest allocation — no fill step
# ⚠  You must overwrite every cell before reading!

arr = np.empty((4, 3), dtype=np.float64)
print(arr)
```
