```python
np.random.seed(42)
arr = np.random.randint(0, 100, size=(5, 5))
print(arr)
arr2 = arr.copy()  # deep copy
arr2[0] = [1, 2, 3, 4, 5]
print('arr2 \n', arr2)
print('arr \n', arr)  # not CHANGED !! -- same as deep copy
```

```python
# 1 create 20 numbers between 0-100 (seed 42) and keep only even numbers
np.random.seed(42)
arr = np.random.randint(0, 100, size=(20))
print(arr)
print('arr[arr % 2 == 0]', arr[arr % 2 == 0])
print()

# 2 create 20 numbers between 0-100 (seed 42) and keep only number smllaer than 40
np.random.seed(42)
arr = np.random.randint(0, 100, size=(20))
print(arr)
print('arr[arr < 40]', arr[arr < 40])
print()

# 3 arr = np.array([10, 20, 30, 99, 40, 50, 99, 60]) keep only numbers not 99
arr = np.array([10, 20, 30, 99, 40, 50, 99, 60])
print('arr[arr != 99]', arr[arr != 99])
```

```python
np.random.seed(42)
arr = np.random.randint(0, 100, size=(10))
print(arr)
print('(arr >= 55) and (arr <= 71) \n', (arr >= 55) & (arr <= 71)) # condetion -> between 51 and 71
print()
print(' (arr >= 55) and (arr <= 71)) \n', arr[(arr >= 55) & (arr <= 71)]) 
print(' (arr >= 55) or (arr <= 71)) \n', arr[(arr >= 55) | (arr <= 71)]) 
```

```python
np.random.seed(42)
arr = np.random.randint(0, 100, size=(5, 5))
print(arr)
print('max in matrix', arr.max())
print('max per column', arr.max(axis=0))
print('max per row', arr.max(axis=1))
print('min', arr.min())
print('arg_max index-max per column #', arr.argmax(axis=0))
print('arg_min index-min per column #', arr.argmin(axis=0))
print('sum per column', arr.sum(axis=0))
print('sum per row', arr.sum(axis=1))
print(arr.shape)
```

```python
#Create a 2-dimensional array with 5 rows and 5 columns. Populate this array with random
#integers that range from 0 to 100. Make sure your seed set to 101.
np.random.seed(101)
arr = np.random.randint(0, 100, size=(5, 5))
print(arr)    
#● Calculate the following statistics for each row:
#○ The sum of the elements
print('sum of the elements', arr.sum())

#○ The minimum value among the elements
print('min', arr.min())

#○ The maximum value among the elements
print('max', arr.max())

#○ The mean (average) of the elements
print('mean', arr.mean())

#○ The median of the elements
print('median', np.median(arr))
```
